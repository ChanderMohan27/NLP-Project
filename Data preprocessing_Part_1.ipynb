### 1.2 Parsing data into the required format

```python
#Count Total values for each tag

print("values in CloseDate")
CloseDate=[]
for elemt in tree.iter(tag = "CloseDate"):
    CloseDate.append(elemt.text)
print(len(CloseDate))
print("Values in salary")
Salary =[]
for elemt in tree.iter(tag = 'Salary'):
    Salary.append(elemt.text)
print(len(Salary))
print("Values in Category")
Category=[]
for elemt in tree.iter(tag = 'Category'):
    Category.append(elemt.text)
print(len(Category))
print("Values in Location")
Location=[]
for elemt in tree.iter(tag = 'Location'):
    Location.append(elemt.text)
print(len(Location))
print("Values in contracttype")
ContractType=[]
for elemt in tree.iter(tag = 'ContractType'):
    ContractType.append(elemt.text)
print(len(ContractType))
print("Values in title")
Title=[]
for elemt in tree.iter(tag = 'Title'):
    Title.append(elemt.text)
print(len(Title))
print("Values in contract time")
ContractTime=[]
for elemt in tree.iter(tag = 'ContractTime'):
    ContractTime.append(elemt.text)
print(len(ContractTime))

print("Values in company")
Company=[]
for elemt in tree.iter(tag = 'Company'):
    Company.append(elemt.text)
print(len(Company))

print("Values in OpenDate ")
OpenDate=[]
for elemt in tree.iter(tag = 'OpenDate'):
    OpenDate.append(elemt.text)
print(len(OpenDate))
print("Value in JobAd")
ID=[]
for elemt in tree.iter(tag = 'JobAd'):
    ID.append(elemt.attrib)
print(len(ID))

```

```python
import pandas as pd
# initialise all the fields to be an empty list
dataDict = {}
dataDict['Source'] = []
dataDict['ID'] = []
dataDict['Title'] = []
dataDict['Location'] = []
dataDict['Company'] = []
dataDict['ContractType'] = []
dataDict['ContractTime'] = []
dataDict['Category'] = []
dataDict['Salary'] = []
dataDict['CloseDate'] = []
dataDict['OpenDate'] = []
#Creating a for loop to create the dataframe
for src in tree.findall("Src"):
    sn = src.attrib["Name"]
    for elem in src.findall("JobAd"):
            Salary = elem.findall("Salary")
            if len(Salary)>0:
                dataDict['Salary'].append(Salary[0].text)
            else:
                dataDict["Salary"].append(None)
            ContractType = elem.findall("ContractType")
            if len(ContractType)>0:
                dataDict['ContractType'].append(elem.findall("ContractType")[0].text)
            else:
                dataDict["ContractType"].append(None)
            ContractTime = elem.findall("ContractTime")
            if len(ContractTime)>0:
                dataDict['ContractTime'].append(elem.findall("ContractTime")[0].text)
            else:
                dataDict["ContractTime"].append(None)
            Company = elem.findall("Company")
            if len(Company)>0:
                dataDict['Company'].append(elem.findall("Company")[0].text)
            else:
                dataDict["Company"].append(None)
            
            dataDict['Source'].append(sn)
            dataDict['ID'].append(elem.attrib["ID"])
            dataDict['OpenDate'].append(elem.findall("OpenDate")[0].text)
            dataDict['CloseDate'].append(elem.findall("CloseDate")[0].text)
            dataDict['Category'].append(elem.findall("Category")[0].text)
            dataDict['Location'].append(elem.findall("Location")[0].text)
            dataDict['Title'].append(elem.findall("Title")[0].text)
        
        
dataDict          
#Creating DataFrame      
df = pd.DataFrame(dataDict)
df
```
