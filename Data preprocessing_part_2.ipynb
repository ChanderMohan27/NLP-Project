### 2. Resolving schema conflicts


```python
#Rename the Column name of dataset2
df1.rename(columns = {'Job Title':'Title', 'Closing':'CloseDate',"Opening":"OpenDate",\
                     "Monthly Payment":"Salary","Type":"ContractTime","Organisation":"Company",\
                      "Full-Time Equivalent (FTE)":"ContractType"}, inplace = True)
df1
```

```python
#Adding a new column 
df1['Source'] = 'www.jobhuntlisting.com'
df1
```

```python
#valuecount for Contract time 
df1["ContractTime"].value_counts()
```

```python
#Changing the datatype 
df1['ContractType'] = df1['ContractType'].replace(1.0, "full_time")
df1['ContractType'] = df1['ContractType'].replace(0.2, "part_time")
df1['ContractType'] = df1['ContractType'].replace(0.6, "part_time")
df1['ContractType'] = df1['ContractType'].replace(0.4, "part_time")
df1['ContractType'] = df1['ContractType'].replace(0.8, "full_time")
```

```python
#valuecount for Contract time 
df1["ContractTime"].value_counts()
```

```python
#replace the value 
df1["ContractTime"].replace({"Fixed Term Contract": "Contract","NaN":"non-specified"}, inplace = True)
df1['ContractTime'] = df1['ContractTime'].fillna("non-specified")
```

```python
#Fill null values with non specified
df1['Company'] = df1['Company'].fillna("non-specified")
```

```python
#add a new column ID 
df1.insert(0, 'ID', range(10000000, 10000000 + len(df1)))
```
